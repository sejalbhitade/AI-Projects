```python
from statsmodels.tsa.stattools import adfuller
```

```python
def adf_test(series,title):
    print('This is Augmented Dickey Fulley Test result in {}'.format(title))
    print()
    print("H0: Data has unit root and is non stationary")
    print('Ha: Data has no unit root and is stationary')
    print('if p>0.05 then accept H0 and p<=0.05 reject H0')
    
    result=adfuller(series.dropna())
    dfout=pd.Series(result[:4],index=['Adf test stat','P-value','#lags','#observations'])
    
    if result[1]>0.05:
        print('------------------------------------------')
        print('Data has unit root and is non stationary')
        
    else:
        print('------------------------------------------')
        print('Data has no unit root and is stationary')
```

```python
adf_test(df.AverageTemperature,'Average Temperature')
```

### ARMA and ARIMA Models

##### Auto correlation is the relation with self 
max strong relation is positive  +1 AND
min weak relation which is negetive -1

##### the value of D will be always 0

```python
from statsmodels.tsa.arima.model import ARIMA
#from dtatsmodel.tsa.arima_model importARMA,ARMAResults,ARIMA,ARIMAResults
```

```python
model= ARIMA(train,order=(0,0,2))
results = model.fit()
results.summary()
```

##### AIC-Akaike information criterion 

##### BIC - Bayesian information crierion

##### the model in which we will get the min value of AIC and BIC then it will be known as best model

```python
import matplotlib.pyplot as plt
```

```python
#getting the ARIMA with number of steps as 36 since we want to make 3 yearprediction and our datais montly sampled 
pred=results.get_forecast(steps=36)
#plotting the observed and forecasted values:
ax1=y['2000':].plot(label='observed')
pred.predicted_mean.plot(ax=ax1,label='ARIMA Forecast',figsize = (15,6), linestyle ='dashed')
#finding the confidence intervals of the forecasts.
pred_ci=pred.conf_int()
ax1.fill_between(pred_ci.index, pred_ci.iloc[:,0],pred_ci.iloc[:,1], color='k', alpha=0.2)
ax1.set_xlabel('Date')
ax1.set_ylabel('Average Temperature')
plt.legend(loc='upper left')
plt.show()
```

### Training the SARIMAX Model:

##### Now we train SARIMAX model using the optional values of the hyperparameters that we found using auto-ARIMA

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
model1=SARIMAX(train,order=(1,1,2), seasonal_order = (1,0,1,12),
                              enforce_stationery =False, enforce_invertibility = False)
fitted_model = model1.fit(maxiter = 200)
print(fitted_model.summary())
```

### SARIMAX Forecasting

```python
#getting the Sarimax with number of steps as 36 since we want to make 3 yearprediction and our datais montly sampled 
pred=fitted_model.get_forecast(steps=36)
#plotting the observed and forecasted values:
ax1=y['2000':].plot(label='observed')
pred.predicted_mean.plot(ax=ax1,label='ARIMA Forecast',figsize = (15,6), linestyle ='dashed')
#finding the confidence intervals of the forecasts.
pred_ci=pred.conf_int()
ax1.fill_between(pred_ci.index, pred_ci.iloc[:,0],pred_ci.iloc[:,1], color='k', alpha=0.2)
ax1.set_xlabel('Date')
ax1.set_ylabel('Average Temperature')
plt.legend(loc='upper left')
plt.show()
```
